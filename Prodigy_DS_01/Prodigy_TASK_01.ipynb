Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("Population.csv")
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Extract relevant columns (country name and population)
country_population = df[['Country Name', '2019']]

# Remove rows with missing values
country_population.dropna(inplace=True)

# Sort by population (optional)
country_population.sort_values(by='2019', ascending=False, inplace=True)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(country_population['Country Name'][:10], country_population['2019'][:10])  # Plotting top 10 countries
plt.xlabel('Country')
plt.ylabel('Population (in billions)')
plt.title('Top 10 Countries by Population in 2019')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

# Filter out the rows corresponding to individual countries (exclude 'World' and other aggregates)
country_population = df[df['Country Name'] != 'World']

# Melt the dataframe to reshape it for plotting
melted_data = country_population.melt(id_vars=['Country Name'], var_name='Year', value_name='Population')

# Convert 'Year' column to numeric
melted_data['Year'] = pd.to_numeric(melted_data['Year'], errors='coerce')

# Plotting the scatter plot
fig = px.scatter(melted_data, x='Year', y='Population', color='Country Name', title='World Population by Country',
                 labels={'Year': 'Year', 'Population': 'Population', 'Country Name': 'Country'})

# Show the plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual data)
population_data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female']
})

# Count the occurrences of each gender
gender_counts = population_data['Gender'].value_counts()

# Plotting the bar chart
plt.bar(gender_counts.index, gender_counts.values)

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in Population')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random ages for a population (assuming ages between 18 and 80)
np.random.seed(42)  # Setting seed for reproducibility
ages = np.random.randint(18, 81, size=1000)

# Plotting the histogram
plt.hist(ages, bins=20, edgecolor='black')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in Population')

# Display the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Number of individuals in the dataset
num_individuals = 10000

# Generate random gender data (assuming equal probability for Male and Female)
gender = np.random.choice(['Male', 'Female'], size=num_individuals)

# Generate random age data (assuming ages between 18 and 80)
age = np.random.randint(18, 81, size=num_individuals)

# Create a DataFrame to store the data
data = pd.DataFrame({'Gender': gender, 'Age': age})

# Plotting the scatter plot
plt.scatter(data['Age'], data['Gender'], alpha=0.5)

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Gender')
plt.title('Gender vs. Age')

# Display the plot
plt.show()
```
